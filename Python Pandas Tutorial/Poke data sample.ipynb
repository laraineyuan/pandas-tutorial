```python
#AGGREGATE STATISTICS (Groupby)

df = pd.read_csv('modified.csv')
df = df.drop(columns=['Legendary'])
df = df.drop(columns=['Unnamed: 0.1'])
df = df.drop(columns=['Unnamed: 0'])
#df.groupby(["Type 1"]).mean(numeric_only=True).sort_values('HP', ascending = False)

df ['count'] = 1

df

df.groupby(['Type 1', 'Type 2']).count()['count']

```

```python
#WORKING WITH LARGE AMOUNTS OF DATA

new_df = pd.DataFrame(columns = df.columns) #create a new datafram that's empty with same column names

for df in pd.read_csv('modified.csv', chunksize=5) : #5 rows will be passed in at a time
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results]) # appends two dataframe together
                    
    #take new dataframe as it go through chunks, append on results, and store it back to new_df
```
